#### 합성곱 신경망의 구조
- 합성곱층 -> 풀링층 -> 밀집층

```python
# 합성곱층
# 합성곱 신경망 : 입력과 가중치를 곱하고 절편을 더하는 선형계산 하지만 밀집층과 달리 각 합성곱은 일부만 사용하여 선형 계산을 수행하여 특성맵을 만듬
# 필터 : 합성곱 신경망의 뉴런개수
# 커널 : 뉴런의 가중치
# 특성맵 : 합성곱 계산을 통해 얻은 출력
from tensorflow import keras
keras.layers.Conv2D(10, kernel_size=(3,3), activation='relu') # 필터개수, 커널크기, 합성곱의 활성화함수
```

```python
# 패딩 : 입력 배열의 주위를 가상의 원소로 채우는 것
# 스트라이드 : 커널의 이동의 크기, 주로 1
keras.layers.Conv2D(10, kernel_size=(3,3), activation='relu',
                   padding='same', strides=1) # padding , same: 입력 주위에 0으로 패딩 하는 것 valid : 패딩없이 입력 배열에서만 합성곱을 하는 것
```

```python
# 풀링층
# 풀링 : 합성곱 층에서 만든 특성 맵의 가로세로 크기를 줄이는 역할
# 풀링의 크기와 스트라이드가 같다
keras.layers.MaxPooling2D(2) # 최대풀링 : 가장 큰 값을 고름
keras.layers.AveragePooling2D(2) # 평균풀링 : 평균값을 계산
```

```python
# 밀집층
# 풀링까지 마친 다차원 데이터를 flatten층으로 1차원으로 만들어 밀집층 계산
```
